```python
import pandas as pd
import numpy as np

career_df = pd.read_csv("../resources/career_recommender.csv")
```

```python
career_df.head(5)
```

## Change the column names to simple names

```python
# change column names
# Define the new column names
new_column_names = {
    'What is your name?': 'name',
    'What is your gender?': 'gender',
    'What was your course in UG?': 'university course',
    'What is your UG specialization? Major Subject (Eg; Mathematics)': 'university specialization',
    'What are your interests?': 'interests',
    'What are your skills ? (Select multiple if necessary)': 'skills',
    'What was the average CGPA or Percentage obtained in under graduation?': 'cwa',
    'Did you do any certification courses additionally?': 'certification',
    'If yes, please specify your certificate course title.': 'certification title',
    'Are you working?': 'employment status',
    'If yes, then what is/was your first Job title in your current field of work? If not applicable, write NA.': 'first job title',
    'Have you done masters after undergraduation? If yes, mention your field of masters.(Eg; Masters in Mathematics)': 'masters degree'
}

career_df.rename(columns=new_column_names, inplace=True)
career_df.tail()
```

```python
career_df.info()
```

```python
career_df['interests'].unique()
```

# drop irrelevant columns

```python
# drop irrelevant columns
career_df.drop(columns=[ 'name', 'first job title', "certification", 'masters degree'], inplace=True)
```

## Encoding Skills Column

```python
# # Split skills into separate columns
career_df["skills"] = career_df.skills.str.replace('\n', '').str.replace('\r', '').str.replace(',', ';').str.split(';')


# create a list of all unique skills
all_skills = set()
for skills in career_df["skills"]:
   if type(skills) is not float:
       all_skills.update(skills)
# # career_df = career_df.explode("skills")
all_skills
```

## Separate Skills to individual columns

```python
# create a new column for each skill
# for skill in all_skills:
#     career_df[skill] = career_df["skills"].apply(lambda x: 20 if type(x) is not float and skill in x  else 0)

skill_columns = pd.DataFrame(0, index=career_df.index, columns=list(all_skills))

for index, row in career_df.iterrows():
    if type(row["skills"]) is not float:
       skills_list = row["skills"]
       skill_columns.loc[index, skills_list] = 20

# concatenate the skill columns with the original dataframe
career_df = pd.concat([career_df, skill_columns], axis=1)

# drop the original skills column
career_df.drop(columns=["skills"], inplace=True)
career_df.head()
```

```python
career_df.Python.value_counts()
```

## Encode Interest Column

```python
career_df.interests.unique()
```

```python
# # Split interest into separate columns
career_df["interests"] = career_df.interests.str.replace('\n', '').str.replace('\r', '').str.replace(',', ';').str.split(';')


# create a list of all unique skills
all_interests = set()
for interests in career_df["interests"]:
   if type(interests) is not float:
       all_interests.update(interests)
# # career_df = career_df.explode("interests")
all_interests
```

```python

interest_columns = pd.DataFrame(0, index=career_df.index, columns=list(all_interests))

for index, row in career_df.iterrows():
    if type(row["interests"]) is not float:
       interests_list = row["interests"]
       interest_columns.loc[index, interests_list] = 10

# concatenate the skill columns with the original dataframe
career_df = pd.concat([career_df, interest_columns], axis=1)

# drop the original interests column
career_df.drop(columns=["interests"], inplace=True)
career_df.head()
```

## Encode Certification Columns

```python
# # Split certification into separate columns
career_df["certification title"] = career_df["certification title"].str.replace('\n', '').str.replace('\r', '').str.replace(',', ';').str.split(';')


# create a list of all unique skills
all_certification = set()
for certification in career_df["certification title"]:
   if type(certification) is not float:
       all_certification.update(certification)
# # career_df = career_df.explode("certification title")
all_certification
```

```python

certification_columns = pd.DataFrame(0, index=career_df.index, columns=list(all_certification))

for index, row in career_df.iterrows():
    if type(row["certification title"]) is not float:
       certification_list = row["certification title"]
       certification_columns.loc[index, certification_list] = 15

# concatenate the skill columns with the original dataframe
career_df = pd.concat([career_df, certification_columns], axis=1)

# drop the original certification title column
career_df.drop(columns=["certification title"], inplace=True)
career_df.head()
```

## Experiment with resulting data

```python
data = career_df.drop(columns=["university course", "university specialization", "employment status", "gender"])
```

```python
# convert cwa to float
data["cwa"] = data["cwa"].str.replace('%', '').astype(float)

# scale down cwa
data["cwa"] = data["cwa"] * 0.1
```
